```python
from plyfile import PlyData, PlyElement
import pcl
```

```python
#load ply file generated by pyrealsense library
loadedPlyFile = pcl.load("./out.ply")

# binary ply to ascii ply
pcl.save(loadedPlyFile, "./pcl.ply")

# binary ply to ascii pcd
pcl.save(loadedPlyFile, "./pcl.pcd")
```
